```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#creating dataset
dataset=np.random.randn(25)
#create another rugplot
sns.rugplot(dataset);
#set x and y axis
x_min=dataset.min()-2
x_max=dataset.max()+2
#100 equally spaced points 
x_axis=np.linspace(x_min,x_max,100)
url='http://https://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'
bandwidth=((4*dataset.std()**5)/(3*len(dataset)))**.2
#create empty kernel list
kernel_list=[]
#plot each basis function
for data_point in dataset:
    kernel=stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    kernel=kernel/kernel.max()
    kernel=kernel*.4
    plt.plot(x_axis,kernel,color='grey',alpha=0.5)
plt.ylim(0,1)
    
```

```python
# to get sum of plot
sum_of_kde=np.sum(kernel_list,axis=0)
#plotfigure
fig=plt.plot(x_axis,sum_of_kde,color='red')
#add initial rugplot
sns.rugplot(dataset,c='red')
#get rid of y tick marks
plt.yticks([])
#set title
plt.suptitle('sum of basis functions')
```
