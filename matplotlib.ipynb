```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL method
plt.plot(x,y)
#plt.show()  if not in jupyter notebook
```

```python
plt.plot(x,y,'-r')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OO method
fig=plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=2)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('1st plot')
axes[1].plot(y,x)
axes[1].set_title('2nd plot')
```

FIGURE SIZE AND DPI

```python
fig=plt.figure(figsize=(7,2))

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(5,2)) # dpi can be saved here too 

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('mypict.png',dpi=200)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3,label='x cubed')
ax.legend(loc=0)  ## ax.legend(loc=(0.1,0.1)) location in form of tuple
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=5,alpha=0.5,linestyle='--',marker='o',markersize=20,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') ## rgb hashcodes for diff color ,alpha for color visibility
#ls for linestle,lw for linewidth
```

```python
fig.savefig('pict1.png')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
